```python
#Import Dependencies  
import pyautogui
import cv2
import os
import numpy as np
import matplotlib.pyplot as plt
import pygetwindow as gw
from ultralytics import YOLO
from ReadWriteMemory import ReadWriteMemory
```

```python
rwm = ReadWriteMemory()

process = rwm.get_process_by_name("deadcells.exe")
process.open()

baseaddress = 0x90000+0x2D7CD8

healthpointer = process.get_pointer(baseaddress, offsets=[0x114, 0x8, 0x0, 0x18, 0x5C, 0x3C, 0x3D4])


#value = process.read(healthpointer)
   
```

```python

#Get game window
gameWindow = gw.getWindowsWithTitle('Dead Cells')[0]

#Define game coordinates
gamePosX = 100
gamePosY = 100
gameWidth = 1285
gameHeight = 760

#move game to coordinates
gameWindow.moveTo(gamePosX,gamePosY)

game_model = YOLO("best.pt")

#loop overframe
while True:
    #get health value
    healthpointer = process.get_pointer(baseaddress, offsets=[0x114, 0x8, 0x0, 0x18, 0x5C, 0x3C, 0x3D4])
    value = process.read(healthpointer)
    
    #take a screenshot
    screen = pyautogui.screenshot(region=(gamePosX+7, gamePosY, gameWidth, gameHeight))
    
    #Make Prediction
    results = game_model.predict(screen)

    #Print Health
    print('Health: ', value)
    
    # Handle the rendering of the images/video
    cv2.imshow('GameCap', results[0].plot())
    
    # Cv2.waitkey
    if cv2.waitKey(1) & 0xFF==ord('q'):
        break
# Close down the frame
cv2.destroyAllWindows()
```
